```r
predict_clusters <- function(atlas, query, .k=5, .eps=.05) {
    knn <- nn2(
        t(singlecellmethods:::cosine_normalize_cpp(atlas$Z_corr, 2)), 
        t(singlecellmethods:::cosine_normalize_cpp(query$Z, 2)), 
        k = .k, eps = .eps
    )
    probs <- as.numeric(matrix(t(prop.table(exp(-knn$nn.dists / .1), 1)), ncol = 1))
    knn_dgc <- sparseMatrix(
        i = as.numeric(rep(seq_len(nrow(knn$nn.idx)), each = ncol(knn$nn.idx))), 
        j = as.numeric(matrix(t(knn$nn.idx), ncol = 1)), 
        x = probs 
    )
    idx_min <- min(as.integer(knn$nn.idx))
    idx_max <- max(as.integer(knn$nn.idx))

    cluster_design <- as(as.matrix(model_matrix(atlas$meta_data[1:idx_max, ], ~ 0 + Cluster_name)), 'dgCMatrix')
    colnames(cluster_design) <- gsub('Cluster_name', '', colnames(cluster_design))
    class_pred_probs <- knn_dgc %*% cluster_design
    class_pred <- colnames(class_pred_probs)[apply(class_pred_probs, 1, which.max)]
    return(class_pred)
}
```

# MASC


```r
## Pseudobulk? 
masc_df <- data.table(query$meta_data)[
    , .N, by = .(Cluster, Subject_Identity)
][] %>% 
    dplyr::select(Cluster, Subject_Identity, N) %>% 
    tidyr::spread(Subject_Identity, N, fill = 0) %>% 
    tidyr::gather(Subject_Identity, N, -Cluster) %>% 
    dplyr::left_join(unique(dplyr::select(query$meta_data, Subject_Identity, Disease_Identity)))


masc_df$Disease_Identity <- factor(masc_df$Disease_Identity, c('Control', 'IPF'))
```

```r
# masc_res <- unique(obj_down$meta_data$Cluster) %>% map(function(cluster_test) {
plan(multicore)
masc_res <- unique(masc_df$Cluster) %>% future_map(function(cluster_test) {
    message(cluster_test)
    data_use <- masc_df %>% ## Pseudobulk
#     data_use <- query$meta_data %>% ## Single cell
        dplyr::mutate(Response = case_when(
            Cluster == cluster_test ~ 1L,
            TRUE ~ 0L
        )) 
    tryCatch({
        glmod <- glFormula(
            Response ~ 1 + Disease_Identity + (1|Subject_Identity),
            data_use, 
            family = 'binomial', 
            weights = data_use$N + 1
        )

        ## Learn model with minimum sigma, for numerical stability 
        devfun <- do.call(mkGlmerDevfun, glmod)
        opt <- optimizeGlmer(devfun, stage = 1)
        devfun <- updateGlmerDevfun(devfun, glmod$reTrms)
        opt <- optimizeGlmer(devfun, stage=2)
        model <- mkMerMod(environment(devfun), opt, glmod$reTrms, fr = glmod$fr)

        tibble(
            Cluster = cluster_test,
            beta = fixef(model)[['Disease_IdentityIPF']] / log(2),
            sigma = sqrt(vcov(model)['Disease_IdentityIPF', 'Disease_IdentityIPF']) / log(2),
            zscore = beta / sigma,
            pvalue = 2 * pnorm(-abs(zscore))
        ) %>% 
        return()

    }, error = function(e) {
        tibble(Cluster = cluster_test, beta = NA, sigma = NA) %>% 
        return()

    })
}) %>% 
    bind_rows() 

#     dplyr::mutate(fdr = p.adjust(pvalue))

# masc_res$fdr <- p.adjust(masc_res$pvalue, 'BH')
```

```r
## Collapse Counts
pb <- presto::collapse_counts(
    adams$counts,
    adams$meta_data,
    c('Subject_Identity', 'Cluster', 'Disease_Identity'),
    keep_n=TRUE, 
    how='sum'
) 
pb$meta_data$logUMI <- log(colSums(pb$counts_mat))
```

```r
genes_test <- rownames(pb$counts_mat)[which(rowSums(pb$counts_mat >= 3) >= 3)]# %>% head(100)
length(genes_test)

genes_test <- intersect(unique(roche$feature), genes_test)
length(genes_test)
```

```r
library(lme4)
library(future)
library(furrr)
library(glue)
library(modelr)
```

```r
## Fit Models
t2 <- system.time({
    suppressWarnings({
        adams$presto <- presto.presto(
#             formula = y~1+InflamScore_beta+(1|Cluster)+(1|LibraryID/Cluster)+(1|Tissue/Cluster)+offset(logUMI), 
            formula = y~1+(1|Cluster)+(1|Subject_Identity/Cluster)+offset(logUMI), 
            design = pb$meta_data, 
            response = pb$counts_mat, 
            size_varname = 'logUMI',
            features = genes_test,
            ncore = 20,
            nsim = 1e3,
            effects_cov = c('Cluster'),
            min_sigma = .5,
            verbose = 1
        ) 
    })
})
```

```r
t2[['elapsed']] / 60
```

## Marginal effects

```r
contrasts_mat <- make_contrast.presto(adams$presto, 'Cluster')
adams$effects <- contrasts.presto(adams$presto, contrasts_mat, one_tailed = TRUE) %>% 
    dplyr::mutate(Cluster = contrast) %>% 
    arrange(pvalue) %>% 
    dplyr::mutate(Cluster_name = Cluster, Cluster = factor(gsub('.*C(\\d+)', '\\1', Cluster)))
```
