### GSEA

```r
future::plan(multiprocess)

do_fgsea <- function(effects, genesets_db, .minSize=1, .nperm=1e4, .nproc=20, verbose=FALSE) {
    fgsea_res <- unique(effects$Cluster) %>% 
        future_map(function(cluster_test) {
        if (verbose) {
            message(cluster_test)
        }
            
        .SD <- effects %>% 
            subset(Cluster == cluster_test)
        stats_vec <- .SD$zscore
        names(stats_vec) <- .SD$feature
        stats_vec <- stats_vec[!is.na(stats_vec)]
        
        suppressWarnings({
            fgsea::fgsea(
                pathways = genesets_db,
                stats = stats_vec,
                nperm = .nperm,
                minSize = .minSize,
                nproc = .nproc
            ) %>% 
                dplyr::mutate(
                    Cluster = cluster_test
                ) %>% 
                dplyr::select(Cluster, everything())            
        })
    }) %>% 
        bind_rows()

}
```

```r
fastq_df <- tibble(
    fname = list.files(fastq_dir, recursive = TRUE, pattern='fastq.\\d.gz$', full.names = TRUE),
    ReadNum = gsub('.*(\\d).gz', 'R\\1', fname),
    LibraryID = gsub('.*?(\\d+_.).fastq.\\d.gz', 'Lib\\1', fname)
)

head(fastq_df)
```

### Matrices

```r
mats <- map(unique(fastq_df$LibraryID), function(.libid) {
    fname <- glue('/data/srlab2/ik936/Roche/data/DSS/KB/{.libid}/abundance.tsv')
    if (file.exists(fname)) {
        res <- fread(fname) %>% 
            cbind(LibraryID = .libid)
    } else {
        return(NULL)
    }
}) 

## Remove failed kallisto runs 
mats <- mats[which(!map_lgl(mats, is.null))]
```

```r
## Put it all together! 
counts_tidy <- bind_rows(mats)

counts <- counts_tidy %>% 
    dplyr::select(target_id, est_counts, LibraryID) %>% 
    tidyr::spread(LibraryID, est_counts) %>% 
    tibble::column_to_rownames('target_id')
```

```r
## Convert to gene symbols 
gtf <- fread(mouse_t2g, header = FALSE) %>% 
    data.frame() %>%     
    tibble::column_to_rownames('V1')
design <- Matrix::sparse.model.matrix(~0+V3, gtf)[rownames(counts), ]
colnames(design) <- gsub('^V3', '', colnames(design))

counts <- (t(design) %*% as.matrix(counts))
```

### metadata 

```r
meta_data <- tibble(
    LibraryID = colnames(counts),
    Code = as.integer(gsub('Lib(\\d+)_.', '\\1', LibraryID)),
    TimePoint = case_when(
        Code == 1 ~ 'D0',
        Code %in% 2:8 ~ 'D2',
        Code %in% 9:15 ~ 'D4',
        Code %in% 16:22 ~ 'D7',
        Code %in% 23:29 ~ 'D9',
        Code %in% 30:35 ~ 'D11',
        Code %in% 36:42 ~ 'D14'
    ),
    TimeInt = as.integer(gsub('D', '', TimePoint)),
    Case = case_when(
        Code %in% c(1, 8, 15, 22, 29, 36) ~ 'SS', 
        TRUE ~ 'DSS'
    ), 
    Cage = case_when(
        Code %in% seq(2, 37, by = 7) ~ 'A',
        Code %in% seq(3, 38, by = 7) ~ 'B',
        Code %in% seq(4, 39, by = 7) ~ 'C',
        Code %in% seq(5, 40, by = 7) ~ 'D',
        Code %in% seq(6, 41, by = 7) ~ 'E',
        Code %in% seq(7, 42, by = 7) ~ 'F',
        Code %in% seq(1, 36, by = 7) ~ 'G'
    ), 
    Depth = colSums(counts)
)
head(meta_data)
```

### LIMMA (peak expression)

P values are really weird when we pool all the data together ... 


```r
library(limma)
g <- unique(effects_mm$feature) %>% intersect(rownames(logcpm))
lres_peaks <- map(paste0('D', c(2, 4, 7, 9, 11, 14)), function(.timepoint) {
    .m <- m %>% subset(TimePoint %in% c('D0', .timepoint)) %>% dplyr::mutate(TimePoint = as.character(TimePoint))
    
    suppressWarnings({
        lmFit(logcpm[g, .m$LibraryID], model.matrix(~1+TimePoint, .m)) %>% 
            eBayes() %>% 
            topTable(number = Inf) %>% 
            tibble::rownames_to_column('feature') %>% 
            cbind(TimePoint = .timepoint) 
    })
    
}) %>% bind_rows()
```

```r
## Get dynamic genes that are upregulated
genes_dynamic_pos <- lres_peaks %>% 
    subset(feature %in% genes_dynamic) %>% 
    subset(logFC > 0 & P.Value < .05) %>% 
    with(unique(feature)) 
length(genes_dynamic_pos)
```
