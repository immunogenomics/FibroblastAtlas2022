## matrices

```r
sumOverRowNames <- function(X) {
    name_factors <- factor(row.names(X))
    res <- presto::sumGroups(X, name_factors)
    row.names(res) <- levels(name_factors)#[1:nrow(res)]
    colnames(res) <- colnames(X)
    return(res)
}
```

```r
read10x_kallisto <- function(run, suffix, ens2hugo, min_counts=500) {
    barcode.loc <- list.files(run, pattern = 'barcodes.txt', full.names = TRUE)
    gene.loc <- list.files(run, pattern = 'features.txt|genes.txt', full.names = TRUE)
    matrix.loc <- list.files(run, pattern = '.mtx', full.names = TRUE)
    
    data <- readMM(file = matrix.loc) %>% as("dgCMatrix") %>% Matrix::t()
    
    cell.names <- readLines(barcode.loc)
    cell.names <- gsub("-1$", "", cell.names)    
    
    if (!missing(suffix)) {
        cell.names <- paste(cell.names, suffix, sep = "_")
    }

    gene.names <- readLines(gene.loc)
    
    row.names(data) <- gene.names
    colnames(data) <- cell.names

    data <- as(data, "dgCMatrix")
    data <- data[, Matrix::colSums(data) >= min_counts]
    row.names(data) <- ens2hugo$symbol[match(row.names(data), ens2hugo$ensembl)]
    data <- data[which(!is.na(row.names(data))), ]
#     return(data)
    return(as(sumOverRowNames(data), "dgCMatrix"))
}
```

```r
plan(multicore)
exprs_list <- future_map(unique(fastq_df$LibraryID), function(LibraryID) {
    message(file.path(kb_outdir, LibraryID, 'counts_unfiltered'))
    read10x_kallisto(
        file.path(kb_outdir, LibraryID, 'counts_unfiltered'), 
        LibraryID, 
        ens2hugo, 
        min_counts = 100
    )
})    
```

# Cluster de novo

```r
exprs_raw <- readRDS('/data/srlab/ik936/Roche/data/cache/exprs_raw_matrix_mouse_cia_2021.rds')
```

```r
mito_genes <- grep("^MT-", rownames(exprs_raw), value = TRUE, ignore.case = TRUE)
meta_data <- tibble(
    CellID = colnames(exprs_raw),
#     RepID = gsub('^.*?\\.(\\d)_(.*$)', '\\1', colnames(exprs_raw)),
#     Status = gsub('^.*?\\.(\\d)_(.*$)', '\\2', colnames(exprs_raw)),
    RepID = gsub('^.*?_(.+)(\\d)', '\\2', colnames(exprs_raw)),
    Status = gsub('^.*?_(.+)(\\d)', '\\1', colnames(exprs_raw)),
    nGene = Matrix::colSums(exprs_raw != 0), 
    nUMI = Matrix::colSums(exprs_raw), 
    percent_mito = Matrix::colSums(exprs_raw[mito_genes, ])/Matrix::colSums(exprs_raw),
    LibraryID = paste0(Status, RepID)
)  
    
```

```r
head(meta_data)
```

```r
table(meta_data$LibraryID)
```

```r
fig.size(5, 8)
set.seed(1)
obj <- create_object(exprs_raw, meta_data) %>% 
    do_harmony(
        'LibraryID', theta=0.5, 
        max.iter.harmony=5, max.iter.cluster=20,
        epsilon.cluster = -Inf, epsilon.harmony = -Inf, 
        plot_convergence = TRUE
    )
```

```r
obj$umap_after <- do_umap(t(obj$Z_corr), .spread=1.0, .min_dist=.8)
```

```r
fig.size(4, 8)
do_scatter(obj$umap_after$embedding, obj$meta_data, 'RepID', quo(Status))
```

```r
obj$adj_after <- obj$umap_after$adj
```

```r
future::plan(multiprocess)
set.seed(3)
t_cluster <- system.time({
    obj <- do_cluster(
        obj, 'adj_after', c(.02, .05, .1),
        append_cols = FALSE, 
        slot_name = 'clusters_df', 
        do_weights = FALSE
    )
})
```
