```r
# library(dplyr)
# library(future)
# library(furrr)
# library(scDblFinder)
# library(scater)
# library(singlecellmethods)
# library(tibble)


scDblFinder_par <- function(counts, library_ids, logcounts=NULL) {
    if (is.null(logcounts)) {
        logcounts <- singlecellmethods::normalizeData(counts, 1e4, 'log')
    }
    logcounts_list <- split(seq_len(ncol(counts)), library_ids) %>% 
        map(function(idx) {
            logcounts[, idx]
        })

    counts_list <- split(seq_len(ncol(counts)), library_ids) %>% 
        map(function(idx) {
            counts[, idx]
        })

    future::plan(multicore)
    future_map2(logcounts_list, counts_list, function(.logcounts, .counts) {
            sce <- SingleCellExperiment(list(counts = .counts,logcounts = .logcounts))
            sce <- scDblFinder(sce, verbose=FALSE)
            as.data.frame(sce@colData) %>% 
                tibble::rownames_to_column('CellID')
        }) %>% 
        bind_rows(.id = 'LibraryID')
}

```

## Presto

```r
## do not test cells that should have been sorted out 
tissue_lineage_df <- list(
    Gut = tibble(Cluster = c('endothelial', 'mural', 'fibroblast', 'plasma', 'glial')),
    SalivaryGland = tibble(Cluster = c('endothelial', 'lymphoid', 'mural', 'fibroblast', 'myeloid', 'epithelial', 'plasma', 'glial')),
    Lung = tibble(Cluster = c('endothelial', 'lymphoid', 'mural', 'fibroblast', 'myeloid', 'epithelial', 'plasma')),
    Synovium = tibble(Cluster = c('endothelial', 'mural', 'fibroblast'))
) %>% bind_rows(.id = 'Tissue')
```

```r
## Collapse Counts
pb <- presto::collapse_counts(
    obj$exprs_raw,
    obj$meta_data %>% dplyr::rename(Cluster = joint_lineage), 
    c('Cluster', 'DonorID', 'Tissue'),
    keep_n=TRUE, 
    how='sum'
) 
pb$meta_data$logUMI <- log(colSums(pb$counts_mat))

pb$meta_data$PrestoID <- colnames(pb$counts_mat)
pb$meta_data <- pb$meta_data %>% 
    dplyr::inner_join(tissue_lineage_df, by = c('Tissue', 'Cluster'))
pb$counts_mat <- pb$counts_mat[, pb$meta_data$PrestoID]
```

```r
head(pb$meta_data)
nrow(pb$meta_data)
```

```r
genes_test <- rownames(pb$counts_mat)[which(rowSums(pb$counts_mat >= 5) >= 10)]
length(genes_test)
```

```r
## ~15 minutes
## Fit Models
suppressWarnings({
    system.time({
        presto_obj <- presto.presto(
            formula = y~1+(1|Cluster)+(1|DonorID/Cluster)+(1|Tissue/Cluster)+logUMI, 
            design = pb$meta_data, 
            response = pb$counts_mat, 
            features = genes_test,
            size_varname = 'logUMI',
            ncore = 24
        ) %>% 
        presto:::effects.presto('Cluster')
    })
})
```

## number conserved vs private

```r
presto_obj <- presto:::effects.presto(presto_obj, c('Cluster', 'Cluster:Tissue'))
```

### Conserved

Conserved in tissues where the cluster actually appears. e.g. fibroblasts should have $\beta_G>0$ for all tissues, but glial cells only for salivary gland and gut. 


```r
fig.size(4, 6)
ngenes_conserved <- data.table(dplyr::inner_join(presto_obj$effect, tissue_lineage_df))[
    , all(fdr < .05 & beta > 0) / .N, by = .(feature, Cluster)
][
    V1 == TRUE, .N, by = Cluster
][order(N)]

ngenes_conserved %>% 
    ggplot(aes(reorder(Cluster, N), N)) + 
        geom_bar(stat = 'identity') + 
        theme_test(base_size = 20) +
        coord_flip() + 
        labs(y = '# conserved markers', x = '') + 
        NULL

ngenes_conserved
```

### Private

Let's define private: 

gene $G$ is private to tissue $T$ *iff*
- $\forall_{S, S\neq T},\beta_{G,S} =0$ 
- $\forall_{S, S\neq T},\beta_{G,T} > \beta_{G,S}$ 



To find such $\{G,T\}$ pairs
- get marker for that cluster
- test $\forall_{S, S\neq T},\beta_{G,T} > \beta_{G,S}$ 
- test $\forall_{S, S\neq T},\beta_{G,S} =0$ 





```r
# x <- data.table(dplyr::inner_join(presto_obj$effect %>% subset(feature %in% c('CCL19', 'VWF', 'THY1')), tissue_lineage_df))
x <- data.table(dplyr::inner_join(presto_obj$effect, tissue_lineage_df))

y <- x[
    ## Candidate genes 
    fdr < .05 & beta > 0, head(.SD[order(-wald)], 1), by = .(Cluster, feature)
] %>% 
dplyr::left_join(x, by = c('Cluster', 'feature'), suffix = c('_fg', '_bg')) %>%
    subset(Tissue_fg != Tissue_bg) %>% 
    dplyr::mutate(
        zscore = (beta_fg - beta_bg) / sqrt((sigma_fg) ^ 2 + (sigma_bg) ^ 2)
#         zscore = (beta_fg - beta_bg) / sqrt((sigma_fg / 20) ^ 2 + (sigma_bg / 20) ^ 2)
    ) %>% 
    dplyr::mutate(
        p_diff = 1 - pnorm(zscore),
        fdr_diff = p.adjust(p_diff, 'BH')
    ) %>% 
    dplyr::mutate(
        filter_eq0 = (fdr_bg > .05) | (fdr_bg < .05 & beta_bg < 0)
    ) 
```

So how many tissue specific genes are there? 

```r
ngenes_private <- data.table(y)[
    , all(filter_eq0 & zscore > 0 & fdr_diff < .05), by = .(Cluster, feature, Tissue_fg)
][V1 == TRUE] %>% 
#     with(table(Cluster, Tissue_fg))
    with(table(Cluster)) %>% 
    data.table()
```

And what are they? 

```r
data.table(y)[
    , all(filter_eq0 & zscore > 0 & fdr_diff < .05), by = .(Cluster, feature, Tissue_fg)
][V1 == TRUE] %>% 
    subset(Cluster == 'fibroblast' & Tissue_fg == 'Synovium') %>% 
    arrange(Tissue_fg)
```
