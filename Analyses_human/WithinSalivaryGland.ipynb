```r
## Collapse Counts
pb <- presto::collapse_counts(
    fib$exprs_raw,
    cbind(
        fib$meta_data, 
#         fib$meta_data %>% dplyr::select(-Cluster), 
        Cluster = fib$clusters_df$res1
    ), 
    c('Cluster', 'DonorID'),
    keep_n=TRUE, 
    how='sum'
) 
pb$meta_data$logUMI <- log(colSums(pb$counts_mat))
pb$meta_data$PrestoID <- colnames(pb$counts_mat)
# pb$meta_data <- pb$meta_data %>% subset(Cluster %in% c('0', '1', '2', '3'))
# pb$meta_data <- pb$meta_data %>% subset(Cluster %in% c('0', '1'))
# pb$counts_mat <- pb$counts_mat[, pb$meta_data$PrestoID]
```

```r
devtools::load_all('/data/srlab/ik936/ReproducibleDGE/presto')
```

```r
library(presto)
library(lme4)
library(future)
library(furrr)

## Fit Models
suppressWarnings({
    system.time({
        presto_obj <- presto.presto(
            formula = y~1+(1|Cluster)+(1|DonorID/Cluster)+logUMI, 
            design = pb$meta_data, 
            response = pb$counts_mat, 
            features = c(
                'TNFSF13B', 'IL7', 'CD34',# 'IL4',
                'LTB1', 'LTB2', 'PDGFRA', 'PDGFRB', 'ICAM1', 
                'VCAM1', 'PDPN', 'CCL19',
                'CCL21', 'CXCL12', 'FAP'
            ),
            size_varname = 'logUMI',
            ncore = 5,
            nsim = 1e3,
            effects_cov = c('Cluster'),
            min_sigma = .5 * log(2),
            verbose = 1
        )
    })
    
})
```

```r
contrasts_mat <- make_contrast.presto(presto_obj, 'Cluster')
presto_obj$effects <- contrasts.presto(presto_obj, contrasts_mat, one_tailed = TRUE) %>% 
    dplyr::mutate(Cluster = contrast) %>% 
    arrange(pvalue)
```

```r
fig.size(5, 8)
presto_obj$effects %>%
#     dplyr::select(feature, Cluster, beta) %>% 
#     tidyr::spread(Cluster, beta) %>% 
    dplyr::select(feature, Cluster, zscore) %>% 
    tidyr::spread(Cluster, zscore) %>% 
    tibble::column_to_rownames('feature') %>% 
    Heatmap()
```

## Redo Presto on labels

```r
## Collapse Counts
pb <- presto::collapse_counts(
    fib$exprs_raw,
    fib$meta_data %>% dplyr::mutate(Cluster = tissue_cell_type), 
    c('Cluster', 'DonorID'),
    keep_n=TRUE, 
    how='sum'
) 
pb$meta_data$logUMI <- log(colSums(pb$counts_mat))

pb$meta_data$PrestoID <- colnames(pb$counts_mat)
pb$meta_data <- pb$meta_data %>% subset(!Cluster %in% c('UNKNOWN'))
pb$counts_mat <- pb$counts_mat[, pb$meta_data$PrestoID]

```

```r
# genes_test <- c(
#     'TNFSF13B', 'IL7', 'CD34', 'IL4',
#     'LTB1', 'LTB2', 'PDGFRA', 'PDGFRB', 'ICAM1', 
#     'VCAM1', 'FAP', 'PDPN', 'CCL19',
#     'CCL21', 'CXCL12'
# )
genes_test <- rownames(pb$counts_mat)[which(rowSums(pb$counts_mat >= 5) >= 10)]
length(genes_test)
```

```r
## Fit Models
suppressWarnings({
    system.time({
        presto_obj <- presto.presto(
            formula = y~1+(1|Cluster)+(1|DonorID/Cluster)+logUMI, 
            design = pb$meta_data, 
            response = pb$counts_mat, 
            features = genes_test,
            size_varname = 'logUMI',
            ncore = 20,
            nsim = 1e3,
            effects_cov = c('Cluster'),
            min_sigma = .5 * log(2),
            verbose = 1
        )
    })
    
})
```

```r
contrasts_mat <- make_contrast.presto(presto_obj, 'Cluster')
presto_obj$effects <- contrasts.presto(presto_obj, contrasts_mat, one_tailed = TRUE) %>% 
    dplyr::mutate(Cluster = contrast) %>% 
    arrange(pvalue)
```
