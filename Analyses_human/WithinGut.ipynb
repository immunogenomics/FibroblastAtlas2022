```r
data_df %>% 
    dplyr::select(DonorID, LibraryID) %>% 
    unique() %>% 
    with(table(DonorID))
```

```r
masc_res <- unique(data_df$Cluster) %>% map(function(cluster_test) {
    data_use <- data_df %>% 
#         subset(!DonorID %in% c('UC3268', 'IBD2809', 'IBD_1272', 'UC1808')) %>% 
#         subset(DonorID %in% c('UC3268', 'IBD2809', 'IBD_1272', 'UC1808')) %>% 
        dplyr::mutate(Response = case_when(
            Cluster == cluster_test ~ 1L,
            TRUE ~ 0L 
        ))
    
#     idx_unpaired <- which(!data_use$DonorID %in% c('UC3268', 'IBD2809', 'IBD_1272', 'UC1808'))
#     data_use$DonorID[idx_unpaired] <- NA
#     data_use <- data_use[-idx_unpaired, ]
    
    glmod <- glFormula(
        ## Donor is allowed to have a baseline inflammation score
        ## Library is not, because each library has only one sample 
        ## By modeling but not contrasting the Donor inflammation score,
        ##     we should model the contribution of increased inflammation within donor 
#         Response ~ 1 + InflamScore_beta + (1|LibraryID) + (InflamScore_beta||DonorID),# + (1|Case),
        Response ~ 1 + percent_cd45_beta3 + (1|LibraryID),# + (1|Case),
#         Response ~ 1 + InflamScore_beta + (1|LibraryID),# + (1|Case),
#         Response ~ 1 + InflamScore_beta + (1|LibraryID),
        data_use, 
        family = 'binomial',
        weights = N
    )
    
    ## Learn model with minimum sigma, for numerical stability 
    devfun <- do.call(mkGlmerDevfun, glmod)
    environment(devfun)$lower <- rep(.1, length(environment(devfun)$lower))
    opt <- optimizeGlmer(devfun, stage = 1)
    devfun <- updateGlmerDevfun(devfun, glmod$reTrms)
    opt <- optimizeGlmer(devfun, stage=2)
    model <- mkMerMod(environment(devfun), opt, glmod$reTrms, fr = glmod$fr)
    
    tibble(
        Cluster = cluster_test,
        beta = fixef(model)[['percent_cd45_beta3']] / log(2),
        sigma = sqrt(vcov(model)['percent_cd45_beta3', 'percent_cd45_beta3']) / log(2),
        pvalue = 2 * (1 - pnorm(abs(beta / sigma)))
    )
}) %>% 
    bind_rows() %>% 
    arrange(-beta/sigma)
```
