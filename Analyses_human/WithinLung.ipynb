### Presto

```r
head(banovich$meta_data)
```

```r
## Collapse Counts
pb <- presto::collapse_counts(
    banovich$exprs_raw, 
    cbind(banovich$meta_data, banovich$clusters_df) %>% dplyr::rename(Cluster = res8, DonorID = Sample_Name), 
#     exprs_raw, 
#     meta_data %>% dplyr::rename(Cluster = celltype, DonorID = Sample_Name), 
    c('Cluster', 'DonorID'),
    keep_n=TRUE, 
    how='sum'
) 
pb$meta_data$logUMI <- log(colSums(pb$counts_mat))
```

```r
genes_test <- rownames(pb$counts_mat)[which(rowSums(pb$counts_mat >= 3) >= 5)]
length(genes_test)
```

```r
# with(pb$meta_data, table(DonorID, Cluster))
```

```r
library(presto)
library(lme4)
library(future)
library(furrr)

## Fit Models
suppressWarnings({
    system.time({
        presto_banovich <- presto.presto(
            formula = y~1+(1|Cluster)+(1|DonorID/Cluster)+logUMI, 
            design = pb$meta_data, 
            response = pb$counts_mat, 
            features = genes_test,
            size_varname = 'logUMI',
            ncore = 20
        ) %>% 
        presto:::effects.presto('Cluster')
    })
})
```

```r
toptable.presto(presto_banovich, max_fdr = .05)
```

```r
set.seed(2)
inflam_scores_df <- readRDS('/data/srlab/ik936/Roche/data/cache/inflam_scores_icdf.rds')
data_df <- pb$meta_data %>% 
    dplyr::left_join(inflam_scores_df) %>% 
    subset(!Case %in% c('LungControl')) %>% 
    subset(!Diagnosis %in% c('NSIP')) %>% 
    identity()


masc_res <- unique(data_df$Cluster) %>% map(function(cluster_test) {
    data_use <- data_df %>% 
        dplyr::mutate(Response = case_when(
            Cluster == cluster_test ~ 1L,
            TRUE ~ 0L
        ))
    
    
    glmod <- glFormula(
        Response ~ 1 + InflamScore_beta + (1|LibraryID),
        data_use, 
        family = 'binomial',
        weights = N
    )
    
    ## Learn model with minimum sigma, for numerical stability 
    devfun <- do.call(mkGlmerDevfun, glmod)
    environment(devfun)$lower <- rep(.1, length(environment(devfun)$lower))
    opt <- optimizeGlmer(devfun, stage = 1)
    devfun <- updateGlmerDevfun(devfun, glmod$reTrms)
    opt <- optimizeGlmer(devfun, stage=2)
    model <- mkMerMod(environment(devfun), opt, glmod$reTrms, fr = glmod$fr)
    
    tibble(
        Cluster = cluster_test,
        beta = fixef(model)[['InflamScore_beta']] / log(2),
        sigma = sqrt(vcov(model)['InflamScore_beta', 'InflamScore_beta']) / log(2),
        pvalue = 2 * (1 - pnorm(abs(beta / sigma)))
    )
}) %>% 
    bind_rows() %>% 
    arrange(pvalue)
```
